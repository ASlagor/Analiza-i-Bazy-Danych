##### Znajdź wszystkich klientów z Polski, Nigerii lub Bangladeszu.

##### Jakie kategorie filmów zostały wypożyczone przez klientów?

```python
result_set = db.execute("""SELECT DISTINCT category.name FROM customer 
                        INNER JOIN rental ON customer.customer_id = rental.customer_id 
                        INNER JOIN inventory ON rental.inventory_id = inventory.inventory_id 
                        INNER JOIN film ON inventory.film_id = film.film_id 
                        INNER JOIN film_category ON film.film_id = film_category.film_id 
                        INNER JOIN category ON film_category.category_id = category.category_id""")
for r in result_set:  
    print(r)
```

##### Znajdź wszystkie kategorie filmów wypożyczonych w Ameryce.

```python
result_set = db.execute("""SELECT DISTINCT category.name, country.country FROM country 
                        INNER JOIN city ON country.country_id = city.country_id  AND country.country in ('United States') 
                        INNER JOIN address ON address.city_id = city.city_id 
                        INNER JOIN customer ON customer.customer_id = address.city_id 
                        INNER JOIN rental ON customer.customer_id = rental.customer_id 
                        INNER JOIN inventory ON rental.inventory_id = inventory.inventory_id 
                        INNER JOIN film ON inventory.film_id = film.film_id 
                        INNER JOIN film_category ON film.film_id = film_category.film_id 
                        INNER JOIN category ON film_category.category_id = category.category_id""")  
for r in result_set:  
    print(r)
```
